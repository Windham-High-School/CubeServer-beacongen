# Beacon Messages @ Interval with Intensity Sweep

```python
# Generated by GPT-4
# Prompted by Joseph R. Freeston (yeah, I'm lazy sometimes)

import csv
import datetime
import io

def create_beacon_csv(start_time, interval, num_messages, body, division, output, encoding, min_intensity, max_intensity, misfire_grace_time):
    with io.StringIO() as csvfile:
        fieldnames = ['time', 'body', 'division', 'output', 'encoding', 'intensity', 'misfire grace time']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()

        current_time = start_time
        intensity_step = (max_intensity - min_intensity) / (num_messages - 1)

        for i in range(num_messages):
            intensity = min_intensity + i * intensity_step
            writer.writerow({
                'time': current_time.strftime('%Y-%m-%dT%H:%M:%S'),
                'body': body,
                'division': division,
                'output': output,
                'encoding': encoding,
                'intensity': round(intensity),
                'misfire grace time': misfire_grace_time
            })
            current_time += interval
        print(csvfile.getvalue())
```

## Parameters:

```python
# Scheduling:

year    = 2023
month   = 4
day     = 4

hour    = 8
minute  = 17
second  = 0

interval = datetime.timedelta(minutes=5)
num_messages = 10
```

```python

# What message?
body = "Hello World"

# Encoding should be "ascii" unless you know what you're doing.
encoding = "ascii"

# Lumen or Nanometer
division = "Lumen"

# Visible or Infrared
output = "Visible"

# Intensity range to sweep
min_intensity = 12
max_intensity = 64

# Grace window to attempt to resend
misfire_grace_time = 20
```

```python
# Generate the CSV data!!
start_time = datetime.datetime(year, month, day, hour, minute, second)
create_beacon_csv(start_time, interval, num_messages, body, division, output, encoding, min_intensity, max_intensity, misfire_grace_time)
```
